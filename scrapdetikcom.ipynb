```python
import requests
import pandas as pd
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup

url = 'https://www.detik.com/search/searchall?query=surabaya&siteid=2&source_kanal=true'
source = requests.get(url)
UserAgent = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
html_text = urlopen(UserAgent)
soup = BeautifulSoup(html_text,"html.parser")

article_list = soup.find_all("article", {"class": "list-content__item"})

nama_penerbit = [tag.find('h2').text.strip() for tag in article_list]
judul_berita = [tag.find('h3').text.strip() for tag in article_list]
tanggal_publikasi = [tag.find('h3').find_next('span').text.strip() if tag.find('h3').find_next('span') else '' for tag in article_list]
link_berita = [tag.find('a')['href'].strip() if tag.find('a') else '' for tag in article_list]

df_berita = pd.DataFrame({
    'Nama Penerbit': nama_penerbit,
    'Judul Berita': judul_berita,
    'Tanggal Publikasi': tanggal_publikasi,
    'Link Berita': link_berita
})

df_berita.index = range(1, len(df_berita) + 1)
print(df_berita)
```

```python
df_berita.to_csv('surabaya_detikcom.csv', index=False, encoding='utf-8')

print("Data berhasil disimpan ke 'surabaya_detikcom.csv'")
```
