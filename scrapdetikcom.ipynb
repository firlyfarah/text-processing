```python
import requests
import pandas as pd
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
```

```python
link = 'https://www.detik.com/search/searchall?query=covid19'
request = requests.get(link)
UserAgent = Request(link, headers={'User-Agent': 'Mozilla/5.0'})
html_text = urlopen(UserAgent)
soup = BeautifulSoup(html_text,"html.parser")
```

```python
daftar = soup.find_all("article", {"class": "list-content__item"})

#menampilkan publish, judul, dan tanggal
publish = [tag.find_previous('h2').text.strip() for tag in daftar]
judul = [element.text.strip().replace("\n", "") for element in daftar]
tanggal = [tag.find_previous('span').text.strip() for tag in daftar]

df = pd.DataFrame({
    'Penerbit':publish,
    'Judul':judul,
    'Tanggal':tanggal

})

print(df)
```

```python
df.to_csv('DetikComCovid19.csv', index=False)
```
